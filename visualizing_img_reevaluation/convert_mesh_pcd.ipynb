```python
import trimesh
import open3d as o3d
import numpy as np

# mesh loading
mesh = trimesh.load("./00010009_d97409455fa543b3a224250f_trimesh_000.xyz.ply.ply")

# sampling from the mesh 
points, face_indices = trimesh.sample.sample_surface(
    mesh,
    8192
)

# create point cloud
pcd = o3d.geometry.PointCloud()

pcd.points = o3d.utility.Vector3dVector(points)

# save as PCD
o3d.io.write_point_cloud("output_cloud.pcd", pcd)

print("finish")
```

```python
# visualize converted point cloud
o3d.visualization.draw_geometries([pcd])
```
